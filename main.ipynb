Fetch Historical Data

```python
import yfinance as yf
from PIL.ImageColor import colormap
```

Downloading 6 months of daily data for reliance

```python
ticker='reliance.ns'
reliance_data=yf.download(ticker, start="2023-01-01", end="2023-07-01")
print(reliance_data)
```

```python
print(reliance_data.shape)
```

Data Preprocessing

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(reliance_data)
plt.show()

```

```python
sns.heatmap(reliance_data.corr(),annot=True)
plt.show()
```

Closing Price Over Time - understanding general price movement

```python
import matplotlib.pyplot as plt
reliance_data['Close'].plot(figsize=(15,5),title='Reliance Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

Daily Trading Volume
Visualize how active trading was, and spot days with unusual volume.

```python
reliance_data['Volume'].plot(figsize=(12, 4), color='orange', title='Daily Trading Volume')
plt.xlabel('Date')
plt.ylabel('Volume (in crores)')
plt.grid(True)
plt.show()
```
